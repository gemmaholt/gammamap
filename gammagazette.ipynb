```python
#center map at hks
boston_coords = [42.371291, -71.122173]

#select columns
gamma_locations = gamma_locations[["Name","Summer","City", "Lat","Long"]]
#load basemap
m = folium.Map(location=boston_coords, zoom_start=1, control_scale=True, 
                 tiles='https://stamen-tiles-{s}.a.ssl.fastly.net/toner-lite/{z}/{x}/{y}{r}.png', 
                 attr='Map tiles by <a href="http://stamen.com">Stamen Design</a>, <a href="http://creativecommons.org/licenses/by/3.0">CC BY 3.0</a> &mdash; Map data &copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors')

# statistical nonsense to get the offsets
scale = 00.0005

spread_cities = ["Boston", "Washington", "New York", "San Francisco"]

update_rows = gamma_locations.City.isin(spread_cities)

gamma_locations.loc[update_rows, ["Lat", "Long"]] += scale*np.random.randn(update_rows.sum().item(),2)

#add data
popups = []
for i, row in gamma_locations.iterrows():    
    html=f'<b>{row.Name}</b> <br><small>{row.Summer}</small>' 
    iframe = branca.element.IFrame(html=html, width=100, height=80)
    popups.append(folium.Popup(iframe, max_width=2650))        
    
marker_cluster = MarkerCluster(locations = gamma_locations[["Lat","Long"]].values,
                               popups = popups).add_to(m)

m
```
